```python
df1[df1.duplicated()].shape 
```

```python
df1.drop_duplicates(inplace = True)
```

```python
df1.shape
```

```python
#Step2: Handling missing values.
```

```python
df1.isnull().sum().sort_values(ascending = False)[:6]
```

```python
df1[['company','agent']] = df1[['company','agent']].fillna(0)
```

```python
df1['children'].unique()
```

```python
df1['children'].fillna(df1['children'].mean(), inplace = True)
```

```python
df1['country'].fillna('others', inplace = True)
```

```python
df1.isnull().sum().sort_values(ascending = False)[:6]
```

```python
df1[df1['adults']+df1['babies']+df1['children'] == 0].shape
```

```python
df1.drop(df1[df1['adults']+df1['babies']+df1['children'] == 0].index, inplace = True)
```

```python
#Step 3: Converting columns to appropriate datatypes.
```

```python
df1[['children', 'company', 'agent']] = df1[['children', 'company', 'agent']].astype('int64')
```

```python
df1['reservation_status_date'] = pd.to_datetime(df1['reservation_status_date'], format='%m/%d/%Y')
```

```python
#Step 4: Adding important columns.
```

```python
df1['total_stay'] = df1['stays_in_weekend_nights']+df1['stays_in_week_nights']
df1['total_people'] = df1['adults']+df1['children']+df1['babies']
```

```python
df1['hotel'].value_counts().plot.pie(explode=[0.05, 0.05], autopct ='%1.1f%%', shadow = True, figsize =(10,9), fontsize = 20)
plt.title('Pie Chart for Most Preferred Hotel', fontsize = 20)
plt.show()
```

```python
group_by_hotel = df1.groupby('hotel')
highest_adr = group_by_hotel['adr'].mean().reset_index()
plt.figure(figsize = (10,8))
ax = sns.barplot(x= highest_adr['hotel'], y= highest_adr['adr'])
ax.set_xlabel("Hotel type", fontsize = 20)
ax.set_ylabel("ADR", fontsize = 20)
ax.set_xticklabels(['City Hotel', 'Resort Hotel'], fontsize = 16)
ax.set_title('Average ADR of each Hotel type', fontsize = 20)
plt.show(ax)
```

```python
adr_vs_total_stay = df1.groupby(['total_stay','adr','hotel']).agg('count').reset_index()
adr_vs_total_stay = adr_vs_total_stay.iloc[:, :3]
adr_vs_total_stay = adr_vs_total_stay.rename(columns = {'is_canceled':'number_of_stays'})
adr_vs_total_stay = adr_vs_total_stay[:18000]
adr_vs_total_stay
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x= 'total_stay', y= 'adr', data= adr_vs_total_stay)
plt.xlabel('Total Stay', fontsize = 16)
plt.ylabel('ADR', fontsize = 16)
plt.title('Relationship between ADR and Total Stay', fontsize = 20)
plt.show()
```

```python
group_by_dc = df1.groupby('distribution_channel')
d1 = pd.DataFrame(round((group_by_dc.size()/df1.shape[0])*100,2)).reset_index().rename(columns = {0: 'Booking_%'})
plt.figure(figsize = (8,8))
data = d1['Booking_%']
labels = d1['distribution_channel']
plt.pie(x=data, autopct="%.2f%%", explode=[0.05]*5, labels=labels, pctdistance=0.5)
plt.title("Booking % by distribution channels", fontsize=14)
```

```python
df1.groupby(['distribution_channel', 'hotel'])['adr'].mean().reset_index()
plt.figure(figsize = (12,6))
sns.barplot(x = 'distribution_channel', y = 'adr', data =df1, hue = 'hotel')
plt.xlabel("Distribution Channel", fontsize = 16)
plt.ylabel("ADR", fontsize = 16)
plt.title('ADR across Distribution Channel', fontsize = 20)
plt.show()
     
```

```python
group_by_dc = df1.groupby('distribution_channel')
d2 = pd.DataFrame(round(group_by_dc['lead_time'].median(),2)).reset_index().rename(columns = {'lead_time': 'median_lead_time'})
plt.figure(figsize = (7,5))
sns.barplot(x = d2['distribution_channel'], y = d2['median_lead_time'])
plt.show()
```

```python
d4 = pd.DataFrame(round((group_by_dc['days_in_waiting_list']).mean(),2)).reset_index().rename(columns = {'days_in_waiting_list': 'avg_waiting_time'})
plt.figure(figsize = (7,5))
sns.barplot(x = d4['distribution_channel'], y = d4['avg_waiting_time'])
plt.show()
```

```python
group_by_dc_hotel = df1.groupby(['distribution_channel', 'hotel'])
d5 = pd.DataFrame(round((group_by_dc_hotel['adr']).agg(np.mean),2)).reset_index().rename(columns = {'adr': 'avg_adr'})
plt.figure(figsize = (7,5))
sns.barplot(x = d5['distribution_channel'], y = d5['avg_adr'], hue = d5['hotel'])
plt.ylim(40,140)
plt.show()
```

```python
bookings_by_months_df = df1.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns = {'hotel':'Counts'})
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
bookings_by_months_df['arrival_date_month'] = pd.Categorical(bookings_by_months_df['arrival_date_month'], categories = months, ordered = True)
bookings_by_months_df = bookings_by_months_df.sort_values('arrival_date_month')
bookings_by_months_df
```
